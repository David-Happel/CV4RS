```python
import numpy as np
from rasterio.windows import Window
from pathlib import Path

imageWidth = 224
imageHeight = 224
window_step = 200

data_dir = "dataset/original/"
data_filename = 'IACS_2018.tif'

out_dir = "dataset/labels/"
out_filename = '{sample}_LABEL.tif'

Path(out_dir).mkdir(parents=True, exist_ok=True)

file_path = data_dir+data_filename

with rasterio.open(file_path) as src:
    print(src.count, src.width, src.height, src.crs) 
    print(src.meta)
    sample_i = 0
    for x in np.arange(0, src.width, window_step):
        for y in np.arange(0, src.height, window_step):              
                
            print("Reading Window:", x,y)
            window = Window(x,y,imageWidth, imageHeight)
            w = src.read(1, window=window)
            show(w)

            # write meta data    
            xform = rasterio.windows.transform(window, src.meta['transform'])    
            meta_d=src.meta.copy()    
            meta_d.update({"height": imageWidth,
                           "width": imageHeight,
                           "transform": xform}) # top left corner
                            

            outfile_path = out_dir+out_filename.format(sample = sample_i)
            with rasterio.open(outfile_path, "w", **meta_d) as dest:
                dest.write_band(1, w)

            sample_i += 1
```
