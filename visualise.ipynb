```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('data/prepared/train/labels.csv')
test_df = pd.read_csv('data/prepared/test/labels.csv')
frames = [train_df, test_df]
df = pd.concat(frames)
df.head(n=5)
df = df.drop(['image_file', 'Unnamed: 0'], axis=1)
df.columns = ['Grassland', 'Winter Wheat', 'Winter Rye', 'Winter Barley', 'Other Winter Cereals', 'Spring Barley', 'Spring Oat', 'Other Spring Cereals', 'Winter Rapeseed', 'Legume', 'Sunflower','Sugar Beet', 'Maize other', 'Maize for grain', 'Potato', 'Strawberry', 'Asparagus', 'Onion', 'Carrot', 'Other leafy Vegetables']
df.info()
df.head()
```

```python
dist = df.sum(axis=0)
```

```python
dist.sort_values(ascending=False, inplace=True)
sns.set(rc={'figure.figsize':(30,15)})
b = sns.barplot(x=dist.index, y=dist)
b.set_ylabel("Sample Count")
b.set_xticklabels(b.get_xticklabels(), rotation=45)
```
